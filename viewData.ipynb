```python
import re

def detect_caption_type(text):
    eng_count = sum([bool(re.search(r"[A-Za-z]", word)) for word in text.split()])
    beng_count = sum([bool(re.search(r"[ঀ-৿]", word)) for word in text.split()])
    if eng_count > 0 and beng_count > 0:
        return "Code-Mixed"
    elif beng_count > 0:
        return "Bengali-Only"
    else:
        return "English-Only"

df["Caption_Type"] = df["Captions"].apply(detect_caption_type)

plt.figure(figsize=(6, 4))
sns.countplot(x="Caption_Type", data=df, order=["Bengali-Only", "English-Only", "Code-Mixed"], palette="pastel")
plt.title("Caption Language Distribution")
plt.xlabel("Caption Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("plots/caption_language.png")
plt.show()
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -- Data Extraction --
img_sizes = []
for fname in df["image_name"]:
    try:
        img = Image.open(os.path.join("Dataset/Memes", fname))
        img_sizes.append(img.size)  # (width, height)
    except:
        continue

widths = [w for w, h in img_sizes]
heights = [h for w, h in img_sizes]

caption_lengths = df["Captions"].apply(lambda x: len(str(x).split()))
sentiment_counts = df["Label_Sentiment"].value_counts()

# -- Generate synthetic stackplot data --
bins = np.linspace(min(widths), max(widths), 10)
stack_vals_w = np.histogram(widths, bins)[0]
stack_vals_h = np.histogram(heights, bins)[0]

# -- Fitbetween plot setup (e.g., token count vs width correlation) --
caption_array = np.array(caption_lengths)
width_array = np.array(widths)
fit = np.polyfit(caption_array, width_array, 1)
fit_line = np.poly1d(fit)(caption_array)
std_dev = np.std(width_array - fit_line)

# -- Plotting --
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.set_style("darkgrid")

# 1. Stackplot: Image Width vs Height Binned Distribution
axes[0, 0].stackplot(bins[:-1], stack_vals_w, stack_vals_h, labels=["Width", "Height"], colors=["#4FC3F7", "#6581FF"])
axes[0, 0].set_title("Stacked Distribution of Image Sizes")
axes[0, 0].set_xlabel("Size Bins (pixels)")
axes[0, 0].set_ylabel("Image Count")
axes[0, 0].legend()

# 2. Fitbetween Plot: Caption length vs Image width
axes[0, 1].scatter(caption_array, width_array, color='skyblue', alpha=0.4)
axes[0, 1].plot(caption_array, fit_line, color='blue', label='Linear Fit')
axes[0, 1].fill_between(caption_array, fit_line - std_dev, fit_line + std_dev, color='blue', alpha=0.3)
axes[0, 1].set_title("FitBetween: Caption Length vs Image Width")
axes[0, 1].set_xlabel("Caption Length (Tokens)")
axes[0, 1].set_ylabel("Image Width")
axes[0, 1].legend()

# 3. Caption Length Distribution (keep one numeric plot)
sns.boxplot(x=caption_lengths, ax=axes[1, 0], color='green')
axes[1, 0].set_title("Boxplot of Caption Lengths")
axes[1, 0].set_xlabel("Token Count")

# 4. Sentiment Pie Chart
axes[1, 1].pie(sentiment_counts.values, labels=sentiment_counts.index,
               autopct='%1.1f%%', colors=["#8BC34A", "#62FF07", "#36F4E4"])
axes[1, 1].set_title("Sentiment Distribution")

plt.suptitle("Dataset Visual Summary", fontsize=16)
plt.tight_layout()
plt.savefig("plots/updated_dashboard.png")
plt.show()
```

```python
import pandas as pd
import re

# Load dataset
# Change filename if needed
df = pd.read_excel("Dataset/multi-sent.xlsx")

# -------------------------------
# Total memes
# -------------------------------
total_memes = len(df)

# -------------------------------
# Sentiment distribution
# -------------------------------
sentiment_counts = df['Label_Sentiment'].value_counts()

positive_count = sentiment_counts.get('positive', 0)
neutral_count  = sentiment_counts.get('neutral', 0)
negative_count = sentiment_counts.get('negative', 0)

# -------------------------------
# Caption language detection
# -------------------------------

# Bengali Unicode range
bengali_pattern = re.compile(r'[\u0980-\u09FF]')
english_pattern = re.compile(r'[A-Za-z]')

def detect_language(text):
    has_bn = bool(bengali_pattern.search(text))
    has_en = bool(english_pattern.search(text))
    
    if has_bn and has_en:
        return "code-mixed"
    elif has_bn:
        return "bengali"
    elif has_en:
        return "english"
    else:
        return "other"

df['caption_lang'] = df['Captions'].astype(str).apply(detect_language)

lang_counts = df['caption_lang'].value_counts()

bengali_captions   = lang_counts.get('bengali', 0)
english_captions   = lang_counts.get('english', 0)
codemixed_captions = lang_counts.get('code-mixed', 0)

# -------------------------------
# Print results (copy to LaTeX)
# -------------------------------
print("MemoSen Dataset Statistics")
print("-" * 30)
print(f"Total memes        : {total_memes}")
print(f"Positive           : {positive_count}")
print(f"Neutral            : {neutral_count}")
print(f"Negative           : {negative_count}")
print(f"Bengali captions   : {bengali_captions}")
print(f"English captions   : {english_captions}")
print(f"Code-mixed captions: {codemixed_captions}")
```
